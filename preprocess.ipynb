```python
import pandas as pd

df = pd.read_csv('14100060.csv')
df.info()
```

```python
df.head()
```

If you look at the value, there are missing data for women's retired age as self-employed in the early days. we just removed it. (assume not enough female enpreneur back then so no need to estimate the value i guess)

```python
df = df[df['STATUS'].isnull()]
df.info()
```

The Coordinate column represent: GEO, retiremnet age, class of work and sex column. It's the premade dummy variable from the statistic canada, so we just removed the rest

```python
df = df[['REF_DATE', 'COORDINATE', 'VALUE']]
df.info()
```

```python
df.head()
```

```python
# Split the 'COORDINATE' column into separate parts
coordinate_parts = df['COORDINATE'].str.split('.', expand=True)

# Assign each part to a new column
df['Country'] = coordinate_parts[0].astype(int)
df['Avg_Median'] = coordinate_parts[1].astype(int)
df['Class_of_Worker'] = coordinate_parts[2].astype(int)
df['Sex'] = coordinate_parts[3].astype(int)

# Drop the original 'COORDINATE' column if it is no longer needed
df.drop('COORDINATE', axis=1, inplace=True)
df.drop('Country', axis=1, inplace=True)
df.info()


```

```python
df.tail()
```

here the first element of the coordinate has been dropped since there's only Canada data. For avg_Median, 1 means average, 2 means median, for class of worker, 1-4 means: total, public, private, self-employed, for sex: 1 is both, 2 is male, 3 is female

```python
# One-hot encode the 'Avg_Median' column
avg_median_dummies = pd.get_dummies(df['Avg_Median'], prefix='Avg_Median')

# One-hot encode the 'Sex' column
sex_dummies = pd.get_dummies(df['Sex'], prefix='Sex')

# One-hot encode the 'Class of worker' column
class_dummies = pd.get_dummies(df['Class_of_Worker'], prefix='Class')

# Concatenate the one-hot encoded columns back to the original DataFrame
df = pd.concat([df, avg_median_dummies, sex_dummies,class_dummies], axis=1)

# Optionally drop the original 'Avg_Median' and 'Sex' columns as they are now encoded
df.drop(['Avg_Median', 'Sex','Class_of_Worker'], axis=1, inplace=True)

df.info()
```
